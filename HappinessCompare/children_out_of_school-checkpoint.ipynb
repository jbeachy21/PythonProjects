```python
import pandas as pd

school_datapath = "Datasets/children_out_of_school_primary.csv"
school_datapath_clean = "Clean_Datasets/children_out_of_school_primary.csv"

school_data = pd.read_csv(school_datapath)
school_data
```

```python
school_data_years = school_data[['country', '2015', '2016','2017','2018','2019']]

school_data_columns = school_data_years.rename(columns={'country':'Country'})

school_data_clean = school_data_columns.dropna()
school_data_clean.reset_index()
school_data_clean = school_data_clean[['Country', '2015','2016','2017','2018','2019']]
school_data_clean
```

```python
school_data_clean.to_csv(school_datapath_clean)
```
